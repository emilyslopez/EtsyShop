## Determing price of traditional and digital art commissions

\* Focusing on nature-focused, painted art

```python
shops_df = pd.DataFrame(shops_dict)
shops_df = shops_df.transpose()
shops_df
```

```python
#Focus on bestsellers only
bestsellers_df = shops_df[shops_df["bestseller"] == 1]
bestsellers_df.replace("NaN", 0, inplace = True)
bestsellers_df
```

```python
reviews = shops_df[["shop_reviews","item_reviews"]] #.sort_values(by = , ascending = False)
reviews
```

```python
reviews_ratio = shops_df["item_reviews"]/shops_df["shop_reviews"]
reviews_ratio = pd.DataFrame(reviews_ratio).rename(columns = {0:"Ratio of Item to Shop Reviews"})
reviews_ratio.sort_values(by = "Ratio of Item to Shop Reviews", ascending = False)
```

Etsy seller **drfrankensara** has the highest proportion of item reviews to overall shop reviews, more than twice the second highest Etsy seller. The bestselling and highest shop reviewed Etsy seller in this list, **LisaNohrenDesigns** is third in this sorted reviews_ratio dataframe.
